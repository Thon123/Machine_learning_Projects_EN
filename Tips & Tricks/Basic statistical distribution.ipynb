# Basic statistical distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

### 1. Normal or Gaussian distribution

```python
from scipy.stats import norm
plt.plot(np.linspace(-5, 5, 1000), norm.pdf(np.linspace(-5, 5, 1000)))
```

### 2. Bernoulli Distribution

Coin flip :), TRU FALSE, only 2 outcomes where one is success

```python
from scipy.stats import bernoulli

# probability of flipping a coin 1 time
p = 0.5
bernouilli_variable = bernoulli(p)

fig, ax = plt.subplots(1, 1)
x = [0, 1]
ax.vlines(x, 0, bernouilli_variable.pmf(x), label='probability')
ax.legend(loc='best', frameon=False)
plt.show()
```

### 3. Binomial Distribution

Bigger brother of Bernoulli, we are counting number of success, we still have the flipping coin.

```python
from scipy.stats import binom

# Probability of outcomes of 'number of successes' when throwing 10 coins
n = 10
p = 0.5
x = np.linspace(0, 10, 11) #number sequence

fig, ax = plt.subplots(1, 1) 
binomial_variable = binom(n, p) #binom function takes probability and number of repetition
ax.vlines(x, 0, binomial_variable.pmf(x), label = 'probability') #
ax.legend(loc='best', frameon=False) 
plt.show()
```

More examples of Binomial

### 4. Poisson Distribution

The Poisson distribution describes a number of events in a fixed time frame.

```python
from scipy.stats import poisson

# Probability of number of events, eg people coming into your store per unit time
lam = 4 # lambda per unit time
x = np.linspace(0, 15, 16)

fig, ax = plt.subplots(1, 1)
poisson_variable = poisson(lam)
ax.vlines(x, 0, poisson_variable.pmf(x), label = 'probability')
ax.legend(loc='best', frameon=False)
plt.show()
```

4 eventy nastanou behem 15 minut, jaka je sance ze nastane X eventu behem 15 minut?

### 5. Exponential Distribution

The exponential distribution describes the waiting time between events.

```python
from scipy.stats import expon

# Probability of time between customers coming in
lam = 4 # event rate of customers coming in per unit time
n_simulated = 10000000

random_waiting_times = expon(scale = 1 / lam).rvs(n_simulated)
pd.Series(random_waiting_times).hist(bins = 20)
```

V průměru každé 4 minuty se něco stane, jaká je šance že se to stane až po 20 minutách? Čím delší doba, tím šance klesá?

### 6. Gamma Distribution

```python

from scipy.stats import gamma

# Probability of time before 10 customers coming in
lam = 4 # event rate of customers coming in per unit time
n_simulated = 10000000

random_waiting_times = gamma(10, scale = 1 / lam).rvs(n_simulated)
pd.Series(random_waiting_times).hist(bins = 20)
```

Čekací doba, jaká je šance že budu čekat X minut když jdu na atrakci, která musí naplnit 10 lidí a dorazí 2 lidi za 2 minuty? Mění se to podle toho, kdy tam dorazím :)

### 7. Weibull Distribution

 It describes a waiting time for one event, if that event becomes more or less likely with time. A clear example would be the life time of a computer. You can wait for a certain time until your computer will be too old and break. The longer you have your computer, the more likely it becomes that it will break. So the probability of failure, or the rate, is not constant.

```python
from scipy.stats import weibull_min

# Probability of time before 10 customers coming in
lam = 4 # event rate of customers coming in per unit time
n_simulated = 10000000
c = 1.2 # larer than 1 so failure rate increases with time
random_waiting_times = weibull_min(c=c, scale = 1 / lam).rvs(n_simulated)
pd.Series(random_waiting_times).hist(bins = 20)
```
